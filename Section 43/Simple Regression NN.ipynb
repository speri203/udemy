# Simple Linear Regression. Minimal example

### Import the relevant libraries

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
sb.set()
```

### Generate random input data to train on

```python
observations = 1000

## The size is number of observations by the number of variables we have. This is
## an example of a basic linear model
xs = np.random.uniform(low=-10, high=10, size=(observations,1))
zs = np.random.uniform(-10,10, (observations,1))

inputs = np.column_stack((xs,zs))
inputs
```

### Create targets we will aim at

```python
noise = np.random.uniform(-1, 1, (observations,1))

targets = 2*xs - 3*zs + 5 + noise
targets.shape
```

### Plot the training data

```python
#
```

### Initialize Variables

```python
init_range = 0.1
weights = np.random.uniform(-init_range, init_range, size=(2,1))
bias = np.random.uniform(-init_range, init_range, size=(1,1))

print(weights)
print(bias)
```

### Specify learning rate

```python
learning_rate = 0.02
```

### Train the Model

```python
for i in range(100):
    outputs = np.dot(inputs, weights) + bias
    deltas = outputs - targets

    loss = np.sum(deltas ** 2) / 2 / observations
    print("Loss: " + str(loss))

    deltas_scaled = deltas / observations

    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    bias = bias - learning_rate * np.sum(deltas_scaled)
```

```python
print(weights, bias)
```

```python
# We print the outputs and the targets in order to see if they have a linear relationship.
# Again, that's not needed. Moreover, in later lectures, that would not even be possible.
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
